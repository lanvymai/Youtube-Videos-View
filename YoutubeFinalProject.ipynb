```python
#Dealing with Missing Data!
#Technique #2- Creating Missing Data Heatmap
cols = df.columns[:30] # first 30 columns
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(palette='colorblind')) 
plt.show()
#or you can also define a palette on your own by giving a hexcolor code as below. 
#Below we are defining two colors yellow and blue for missing and non-missing data.
colors = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
#now we will use this palette in heatmap() of sns
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colors))
plt.show()
```

```python
#Dealing with Missing Data!
#Technique 2: Creating Missing Data Distribution Plot
# first create missing indicator for columns with missing data
for col in df1.columns:
    missing = df1[col].isnull() #missing will create a series with all the columns with null values
    num_missing = missing.sum() #this will sum up all the rows that have nulls   
    #now we want to store the columns which have missing data
    if num_missing > 0:  #this condition is still in the for loop.
        print('created missing indicator for: {}'.format(col)) #prints all columns which have missing data
        df1['{}_ismissing'.format(col)] = missing 
        
        #this will take values from missing series one by one and create a column for them
```

```python
df1
```

```python
# then based on the indicator, plot the histogram of missing values
ismissing_cols = [col for col in df1.columns if 'ismissing' in col] #searching for ismissing in col using list 
#comprehension and storing it in ismissing_cols
df1['num_missing'] = df1[ismissing_cols].sum(axis=1) #create a new column 'num_missing' 

#using displot() from seaborn to plot a distributional summary
sns.displot(data=df1,x='num_missing',kde=True)
plt.show()

#You can also plot the histogram using below command. try it!
df1['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index', y='num_missing')
```

```python
#Technique 3 to check missing data- Calculate percent of missing data in the columns

for col in df1.columns:
    pct_missing = df1[col].isnull().mean() #calculating % of missing values. calling numpy (np)
    print('{} - {}%'.format(col, round(pct_missing*100))) #lookup format function
print(pct_missing)

#looking at the percent values you can decide which columns to delete and which to keep
```

```python
#Solution 3: Impute the missing data
# replace missing values with the median.
med = df1['comment'].median() #calculate the median first
print(med)
df1['comment'] = df1['comment'].fillna(med) #create a new column with the calculated median
```

```python
# Getting started with regression
import patsy                           # provides a syntax for specifying models  
import statsmodels.api as sm           # provides statistical models like ols, gmm, anova, etc...
import statsmodels.formula.api as smf  # provides a way to directly spec models from formulas

```

```python
import ols.formula.api as smf  # provides a way to directly spec models from formulas
```

```python
view_model3 = smf.ols('ln_view ~ category_Howto_Style + category_People_Blogs + category_Science_Technology + ln_comment + ln_duration + ln_like + quality_hd + caption_True', data=df4) 
view_model2 = smf.ols('ln_view ~ category_Howto_Style + category_People_Blogs + category_Science_Technology + ln_duration + ln_like + quality_hd + caption_True', data=df4) 
view_model1 = smf.ols('ln_view ~ category_Howto_Style + category_People_Blogs + category_Science_Technology', data=df4) 
```

```python
results1 = view_model1.fit()
results2 = view_model2.fit()
results3 = view_model3.fit()
```

```python
print(results1.summary())
```

```python
print(results2.summary())
```

```python
print(results3.summary())
```

```python
results1.resid.plot()
```

```python
results2.resid.plot()
```

```python
results3.resid.plot()
```

```python
from statsmodels.iolib.summary2 import summary_col #create regression table for your report

table = summary_col(
    [results1, results2, results3], #results of the fitted models
    model_names = ['view_model1', 'view_model2','view_model3'], #models fitted above
    stars=True, #level of significance
    regressor_order = ['Intercept', 'How To & Styles category', 'People & Blogs category', 'Science & Technology category',
                       'Log of Comment','Squaroot of Duration','Log of Like','Quality','Caption'], #order of explanatory vars in the table
    float_format='%0.2f', 
    drop_omitted = False #
    )

table.add_title('OLS regression model with transformed data')
#table.add_text('any other table info...') #add other table info to be printed with the table

print(table.as_text())

# The 'w' is for 'write'.
fout = open('table1.txt', 'w')
fout.write(table.as_text()) #can export as text, latex or html
fout.close()
```
